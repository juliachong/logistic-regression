```python
import pandas as pd
df = pd.read_csv('adult.zip')
df.dropna(inplace=True)
```

```python
binary_col = []
for i in df['income']:
    if i == '<=50K':
        binary = 0
        binary_col.append(binary)
    if i == '>50K':
        binary1 = 1
        binary_col.append(binary1)
```

```python
df['binary_col'] = binary_col
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(8,6))
plt.scatter(df['hours.per.week'], df['binary_col'])
plt.show()
```
